```python
maleCount = 0
femaleCount = 0
for index, row in patient_data.iterrows():
    if(row['male'] == 1):
        maleCount = maleCount + 1;
    else:
        femaleCount = femaleCount + 1
size = [maleCount, femaleCount]
labels = ['male', 'female']
plt.pie(size, labels = labels, autopct='%1.1f%%')
plt.axis('equal')
```
